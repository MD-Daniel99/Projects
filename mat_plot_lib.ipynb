```python

x_values = list(range(1, 1001))
y_values = [x**2 for x in x_values]

fig, ax = plt.subplots() # Ф-ция subplots() генерирует одну или несколько поддиаграмм на одной диаграмме. ax представляет одну диаграмму на рисунке, fig - одну или несколько
ax.scatter(x_values, y_values, c = y_values, cmap = plt.cm.Reds, s = 100) # Нанесение на диаграмму отдельных точек на основе характеристик

ax.plot(x_values, y_values, linewidth=3) # plot - графическое представление для заданных чисел
ax.set_title("Cubes", fontsize = 24)
ax.set_xlabel("Value", fontsize = 14)
ax.set_ylabel("Cubes of value" ,fontsize = 14)

ax.tick_params(axis = 'both', which = 'major', labelsize = 14) # оформление делений на осях
ax.axis([0, 1001, 0, 1000001])

plt.show() # Вывод окна просмотра
#plt.savefig('squares_plot.png', bbox_inches = 'tight')
```
